## 📚 חלק 1: ייבוא ספריות

### הסבר על הספריות:
| ספרייה | תפקיד |
|--------|--------|
| `pandas` | עבודה עם טבלאות נתונים (DataFrame) |
| `numpy` | חישובים מתמטיים על מערכים |
| `KNeighborsClassifier` | האלגוריתם KNN עצמו |
| `accuracy_score` | חישוב אחוז דיוק |
| `train_test_split` | חלוקת נתונים לאימון ובדיקה |
| `StandardScaler` | נרמול מאפיינים |

</div>

```python
# ייבוא ספריות בסיסיות
import pandas as pd                                    # עבודה עם טבלאות
import numpy as np                                     # חישובים מתמטיים
import matplotlib.pyplot as plt                        # גרפים
import seaborn as sns                                  # גרפים מתקדמים

# ייבוא מ-scikit-learn
from sklearn.neighbors import KNeighborsClassifier     # אלגוריתם KNN
from sklearn.metrics import accuracy_score             # מדד דיוק
from sklearn.metrics import classification_report      # דוח מפורט
from sklearn.metrics import confusion_matrix           # מטריצת בלבול
from sklearn.model_selection import train_test_split   # חלוקת נתונים
from sklearn.preprocessing import StandardScaler       # נרמול

print("כל הספריות נטענו בהצלחה!")
```

<div dir="rtl" style="text-align: right;">

## 📊 חלק 2: טעינת נתונים

### הפקודה: `pd.read_csv(path)`
- **מה עושה:** קוראת קובץ CSV ויוצרת DataFrame
- **פרמטר:** נתיב לקובץ (יכול להיות URL)
- **מחזירה:** DataFrame (טבלה)

</div>

```python
# טעינת מאגר הנתונים Iris
url = 'https://raw.githubusercontent.com/arielb30/datasets/main/Iris.csv'
df = pd.read_csv(url)

# הצגת 5 שורות ראשונות
print("5 השורות הראשונות:")
df.head()
```

```python
# מידע על המאגר
print("מידע על המאגר:")
print(f"מספר שורות: {df.shape[0]}")
print(f"מספר עמודות: {df.shape[1]}")
print(f"\nשמות העמודות: {list(df.columns)}")
print(f"\nסוגי הפרחים: {df['Species'].unique()}")
```

<div dir="rtl" style="text-align: right;">

---
## 🔧 חלק 3: הכנת נתונים

### שלב 3.1: הסרת עמודות מיותרות
הפקודה: `df.drop(columns, axis=1, inplace=True)`
- **columns:** רשימת עמודות להסרה
- **axis=1:** ציר העמודות (0=שורות, 1=עמודות)
- **inplace=True:** שינוי ה-DataFrame המקורי

</div>

```python
# הסרת עמודת Id (לא רלוונטית לסיווג)
df.drop('Id', axis=1, inplace=True)

print("עמודות לאחר הסרה:")
print(list(df.columns))
```

<div dir="rtl" style="text-align: right;">

### שלב 3.2: ערבוב הנתונים
הפקודה: `df.sample(frac=1)`
- **frac=1:** לקחת 100% מהנתונים בסדר אקראי
- **למה חשוב?** הנתונים מסודרים לפי סוג - נרצה לערבב

</div>

```python
# ערבוב הנתונים
df = df.sample(frac=1, random_state=42)

# איפוס האינדקס
df.reset_index(drop=True, inplace=True)

print("לאחר ערבוב - 5 שורות ראשונות:")
df.head()
```

<div dir="rtl" style="text-align: right;">

### שלב 3.3: הפרדת מאפיינים (X) ומטרה (y)

| משתנה | תוכן | צורה |
|-------|------|------|
| **X** | המאפיינים (Features) | מטריצה (n_samples, n_features) |
| **y** | התוויות (Labels) | וקטור (n_samples,) |

</div>

```python
# הפרדת מאפיינים (X) ומטרה (y)

# X = כל העמודות חוץ מ-Species, המרה למערך numpy
X = df.drop(['Species'], axis=1).to_numpy()

# y = רק עמודת Species, המרה למערך numpy  
y = df['Species'].to_numpy()

print(f"צורת X (מאפיינים): {X.shape}")
print(f"צורת y (תוויות): {y.shape}")
print(f"\nדוגמה - שורה ראשונה של X: {X[0]}")
print(f"דוגמה - תווית ראשונה של y: {y[0]}")
```

<div dir="rtl" style="text-align: right;">

---
## 🤖 חלק 4: יצירת מודל KNN ואימון בסיסי

### הפקודה: `KNeighborsClassifier(n_neighbors=K)`
- **מה עושה:** יוצרת אובייקט מסווג KNN
- **n_neighbors:** מספר השכנים (K)

### הפקודה: `model.fit(X, y)`
- **מה עושה:** "מאמנת" את המודל (בפועל: שומרת את הנתונים)
- **X:** מטריצת המאפיינים
- **y:** וקטור התוויות

</div>

```python
# יצירת מודל KNN עם K=3
knn = KNeighborsClassifier(n_neighbors=3)

# אימון המודל
knn.fit(X, y)

print("המודל אומן בהצלחה!")
print(f"מספר שכנים: {knn.n_neighbors}")
```

<div dir="rtl" style="text-align: right;">

---
## 🔮 חלק 5: חיזוי (Prediction)

### הפקודה: `model.predict(X_new)`
- **מה עושה:** מחזירה חיזוי לכל דוגמה
- **X_new:** מאפיינים של דוגמאות חדשות (מטריצה 2D!)
- **מחזירה:** מערך של תוויות

</div>

```python
# חיזוי על 3 דוגמאות בדויות
# שים לב: חייבים להעביר מערך דו-ממדי [[...], [...], [...]]

new_flowers = [
    [1, 1, 1, 1],      # פרח עם ערכים קטנים
    [4, 4, 4, 4],      # פרח עם ערכים בינוניים  
    [5, 2, 3, 1]       # פרח אחר
]

predictions = knn.predict(new_flowers)
print("חיזויים לפרחים החדשים:")
print(predictions)
```

```python
# חיזוי על כל הנתונים
y_pred = knn.predict(X)

print(f"צורת וקטור החיזויים: {y_pred.shape}")
print(f"\n5 חיזויים ראשונים: {y_pred[:5]}")
print(f"5 תוויות אמיתיות: {y[:5]}")
```

<div dir="rtl" style="text-align: right;">

---
## 📏 חלק 6: הערכת דיוק (Accuracy)

### שיטה 1: `model.score(X, y)`
- **מה עושה:** מבצעת predict ואז מחשבת accuracy
- **מחזירה:** מספר בין 0 ל-1

### שיטה 2: `accuracy_score(y_true, y_pred)`
- **מה עושה:** משווה תוויות אמיתיות לחזויות
- **y_true:** התוויות האמיתיות
- **y_pred:** התוויות שהמודל חזה

</div>

```python
# שיטה 1: שימוש ב-score
accuracy1 = knn.score(X=X, y=y)
print(f"דיוק (שיטה 1 - score): {accuracy1:.2%}")

# שיטה 2: שימוש ב-accuracy_score
accuracy2 = accuracy_score(y_true=y, y_pred=y_pred)
print(f"דיוק (שיטה 2 - accuracy_score): {accuracy2:.2%}")
```

<div dir="rtl" style="text-align: right;">

---
## ⚠️ חלק 7: חלוקה נכונה - Train/Test Split

### הפקודה: `train_test_split(X, y, test_size, random_state)`
- **X, y:** המאפיינים והתוויות
- **test_size:** אחוז לבדיקה (0.2 = 20%)
- **random_state:** זרע לחלוקה אקראית (לשחזור)
- **מחזירה:** X_train, X_test, y_train, y_test

### למה זה חשוב?
אם נבדוק על אותם נתונים שאימנו - נקבל תוצאה אופטימית מדי!

</div>

```python
# חלוקה ל-80% אימון, 20% בדיקה
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,      # 20% לבדיקה
    random_state=42     # לשחזור
)

print(f"נתוני אימון: {X_train.shape[0]} דוגמאות")
print(f"נתוני בדיקה: {X_test.shape[0]} דוגמאות")
```

```python
# אימון על Train, בדיקה על Test
knn2 = KNeighborsClassifier(n_neighbors=3)
knn2.fit(X_train, y_train)

# הערכה על נתוני הבדיקה
test_accuracy = knn2.score(X_test, y_test)
print(f"דיוק על נתוני בדיקה: {test_accuracy:.2%}")
```

<div dir="rtl" style="text-align: right;">

---
## 📐 חלק 8: נרמול (Scaling) - חשוב ל-KNN!

### הפקודה: `StandardScaler()`
- **מה עושה:** מביאה כל מאפיין לממוצע 0 וסטיית תקן 1
- **למה חשוב?** KNN מחשב מרחקים - מאפיינים בסקאלות שונות ישפיעו באופן לא שווה

### `scaler.fit_transform(X_train)` vs `scaler.transform(X_test)`
- **fit_transform:** לומד את הפרמטרים (ממוצע, סטיית תקן) וגם מבצע טרנספורמציה
- **transform:** רק מבצע טרנספורמציה עם פרמטרים שכבר נלמדו

### ⚠️ חשוב: fit רק על Train!

</div>

```python
# יצירת אובייקט הנרמול
scaler = StandardScaler()

# fit_transform על נתוני אימון בלבד!
X_train_scaled = scaler.fit_transform(X_train)

# transform בלבד על נתוני בדיקה
X_test_scaled = scaler.transform(X_test)

print("לפני נרמול (Train):")
print(f"  ממוצע: {X_train.mean(axis=0).round(2)}")
print(f"  סטיית תקן: {X_train.std(axis=0).round(2)}")

print("\nאחרי נרמול (Train):")
print(f"  ממוצע: {X_train_scaled.mean(axis=0).round(3)}")
print(f"  סטיית תקן: {X_train_scaled.std(axis=0).round(3)}")
```

```python
# אימון על נתונים מנורמלים
knn_scaled = KNeighborsClassifier(n_neighbors=3)
knn_scaled.fit(X_train_scaled, y_train)

# הערכה
scaled_accuracy = knn_scaled.score(X_test_scaled, y_test)
print(f"דיוק עם נרמול: {scaled_accuracy:.2%}")
print(f"דיוק ללא נרמול: {test_accuracy:.2%}")
```

<div dir="rtl" style="text-align: right;">

---
## 🔍 חלק 9: שיטת המרפק - מציאת K אופטימלי

### העיקרון:
1. נריץ KNN עם ערכי K שונים (1 עד 30)
2. נחשב את שיעור השגיאה לכל K
3. נבחר את K בנקודת "המרפק" - שם השגיאה מפסיקה לרדת משמעותית

</div>

```python
# שיטת המרפק
k_range = range(1, 31)  # K מ-1 עד 30
error_rates = []

for k in k_range:
    # יצירת מודל עם K נוכחי
    knn_temp = KNeighborsClassifier(n_neighbors=k)
    
    # אימון
    knn_temp.fit(X_train_scaled, y_train)
    
    # חיזוי
    y_pred_temp = knn_temp.predict(X_test_scaled)
    
    # חישוב שגיאה (1 - דיוק)
    error = 1 - accuracy_score(y_test, y_pred_temp)
    error_rates.append(error)

# מציאת K הטוב ביותר
best_k = list(k_range)[np.argmin(error_rates)]
print(f"K האופטימלי: {best_k}")
print(f"שיעור שגיאה מינימלי: {min(error_rates):.2%}")
```

```python
# ציור גרף המרפק
plt.figure(figsize=(10, 5))
plt.plot(list(k_range), error_rates, 'o-', color='red', linewidth=2)
plt.axvline(x=best_k, color='orange', linestyle='--', label=f'Best K = {best_k}')
plt.xlabel('K (Number of Neighbors)')
plt.ylabel('Error Rate')
plt.title('Elbow Method for Optimal K')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

<div dir="rtl" style="text-align: right;">

---
## 📋 חלק 10: דוח סיווג מלא

### הפקודה: `classification_report(y_true, y_pred)`
מחזירה דוח עם:
- **Precision:** מתוך מה שחזינו כמחלקה X, כמה באמת X
- **Recall:** מתוך כל המחלקה X, כמה מצאנו
- **F1-score:** ממוצע הרמוני של Precision ו-Recall

</div>

```python
# אימון מודל סופי עם K אופטימלי
final_knn = KNeighborsClassifier(n_neighbors=best_k)
final_knn.fit(X_train_scaled, y_train)

# חיזוי סופי
y_final_pred = final_knn.predict(X_test_scaled)

# דוח סיווג
print("דוח סיווג מלא:")
print("=" * 60)
print(classification_report(y_test, y_final_pred))
```

```python
# מטריצת בלבול
cm = confusion_matrix(y_test, y_final_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=np.unique(y),
            yticklabels=np.unique(y))
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

## 📝 סיכום הפקודות העיקריות

| פקודה | תפקיד | דוגמה |
|-------|--------|--------|
| `pd.read_csv(path)` | קריאת קובץ CSV | `df = pd.read_csv('data.csv')` |
| `df.drop(col, axis=1)` | הסרת עמודה | `df.drop('Id', axis=1)` |
| `df.to_numpy()` | המרה למערך | `X = df.to_numpy()` |
| `train_test_split()` | חלוקת נתונים | `X_train, X_test, y_train, y_test = ...` |
| `StandardScaler()` | יצירת מנרמל | `scaler = StandardScaler()` |
| `scaler.fit_transform(X)` | למידה + נרמול | `X_scaled = scaler.fit_transform(X)` |
| `scaler.transform(X)` | נרמול בלבד | `X_test_scaled = scaler.transform(X_test)` |
| `KNeighborsClassifier(n_neighbors=K)` | יצירת מודל | `knn = KNeighborsClassifier(n_neighbors=5)` |
| `model.fit(X, y)` | אימון | `knn.fit(X_train, y_train)` |
| `model.predict(X)` | חיזוי | `y_pred = knn.predict(X_test)` |
| `model.score(X, y)` | חישוב דיוק | `acc = knn.score(X_test, y_test)` |
| `accuracy_score(y_true, y_pred)` | חישוב דיוק | `acc = accuracy_score(y_test, y_pred)` |
| `classification_report()` | דוח מלא | `print(classification_report(y_test, y_pred))` |
