## 💡 הרעיון המרכזי של SVM

**SVM מחפש את הקו (או המישור) שמפריד הכי טוב בין שתי קבוצות.**

### מה זה "הכי טוב"?
- הקו שנמצא **הכי רחוק** מהנקודות של שתי הקבוצות
- המרחק הזה נקרא **Margin** (שוליים)
- SVM **ממקסם את ה-Margin**

### למה זה חשוב?
- קו עם Margin גדול = יותר "בטחון" בסיווג
- פחות רגיש לנקודות חדשות

</div>

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# ויזואליזציה של הרעיון של Margin

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

np.random.seed(42)
# נתונים פשוטים
class_a = np.random.randn(20, 2) + np.array([2, 2])
class_b = np.random.randn(20, 2) + np.array([-2, -2])

for ax in axes:
    ax.scatter(class_a[:, 0], class_a[:, 1], c='blue', s=100, label='Class A', edgecolors='black')
    ax.scatter(class_b[:, 0], class_b[:, 1], c='red', s=100, label='Class B', edgecolors='black')
    ax.set_xlim(-5, 5)
    ax.set_ylim(-5, 5)
    ax.set_aspect('equal')
    ax.legend()

# קו 1 - לא טוב (קרוב לקבוצה אחת)
x_line = np.linspace(-5, 5, 100)
axes[0].plot(x_line, -x_line - 1.5, 'g-', linewidth=2)
axes[0].fill_between(x_line, -x_line - 1.5, -x_line - 2, alpha=0.3, color='green')
axes[0].set_title('❌ Margin קטן\nקו קרוב מדי לקבוצה אחת', fontsize=12)

# קו 2 - לא טוב (בכלל לא מפריד טוב)
axes[1].plot(x_line, -x_line + 1, 'g-', linewidth=2)
axes[1].fill_between(x_line, -x_line + 0.5, -x_line + 1.5, alpha=0.3, color='green')
axes[1].set_title('❌ Margin קטן\nלא באמצע', fontsize=12)

# קו 3 - מושלם (באמצע עם Margin מקסימלי)
axes[2].plot(x_line, -x_line, 'g-', linewidth=3)
axes[2].fill_between(x_line, -x_line - 1.5, -x_line + 1.5, alpha=0.3, color='green')
axes[2].plot(x_line, -x_line - 1.5, 'g--', linewidth=1)
axes[2].plot(x_line, -x_line + 1.5, 'g--', linewidth=1)
axes[2].set_title('✅ Margin מקסימלי!\nהקו הטוב ביותר', fontsize=12)

plt.suptitle('SVM מחפש את הקו עם ה-Margin הגדול ביותר', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
````

<div dir="rtl" style="text-align: right; background-color: #fff3e0; padding: 20px; border-radius: 10px;">

## 📌 Support Vectors - מה זה?

**Support Vectors** = הנקודות שנמצאות **הכי קרוב** לקו ההפרדה.

### למה הן חשובות?
- הן "תומכות" בקו - קובעות את מיקומו
- רק הנקודות האלה משפיעות על המודל!
- כל שאר הנקודות יכולות להיעלם - הקו לא ישתנה

### זה מה שהופך את SVM למיוחד:
- אלגוריתמים אחרים משתמשים בכל הנקודות
- SVM משתמש רק ב-Support Vectors

</div>

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# הדגמת Support Vectors

from sklearn.svm import SVC

# יצירת נתונים
np.random.seed(42)
X = np.vstack([
    np.random.randn(30, 2) + [2, 2],
    np.random.randn(30, 2) + [-2, -2]
])
y = np.array([0] * 30 + [1] * 30)

# אימון SVM
svm_model = SVC(kernel='linear', C=1)
svm_model.fit(X, y)

# ויזואליזציה
fig, ax = plt.subplots(figsize=(10, 8))

# נקודות רגילות
ax.scatter(X[y==0, 0], X[y==0, 1], c='blue', s=100, label='Class 0', alpha=0.6)
ax.scatter(X[y==1, 0], X[y==1, 1], c='red', s=100, label='Class 1', alpha=0.6)

# Support Vectors - מודגשים!
sv = svm_model.support_vectors_
ax.scatter(sv[:, 0], sv[:, 1], s=300, facecolors='none', edgecolors='black', linewidths=3, label='Support Vectors')

# קו ההפרדה
w = svm_model.coef_[0]
b = svm_model.intercept_[0]
x_line = np.linspace(-5, 5, 100)
decision_boundary = -(w[0] * x_line + b) / w[1]
margin = 1 / np.sqrt(np.sum(w**2))

ax.plot(x_line, decision_boundary, 'g-', linewidth=2, label='Decision Boundary')
ax.plot(x_line, decision_boundary - margin, 'g--', linewidth=1)
ax.plot(x_line, decision_boundary + margin, 'g--', linewidth=1)
ax.fill_between(x_line, decision_boundary - margin, decision_boundary + margin, alpha=0.2, color='green')

ax.set_xlim(-5, 6)
ax.set_ylim(-5, 6)
ax.legend(loc='upper right')
ax.set_title(f'SVM עם {len(sv)} Support Vectors (מסומנים בעיגול שחור)', fontsize=14)
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')

plt.tight_layout()
plt.show()

print(f"📊 מספר Support Vectors: {len(sv)}")
print(f"📊 מתוך סה״כ נקודות: {len(X)}")
print(f"💡 רק {len(sv)/len(X)*100:.1f}% מהנקודות קובעות את המודל!")
````

## ⚠️ הבעיה: נתונים לא לינאריים

לפעמים **אי אפשר להפריד** את הנתונים עם קו ישר!

**דוגמה:** נקודות מסודרות במעגלים - אין קו שיפריד ביניהן.

</div>

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# נתונים לא לינאריים - מעגלים

from sklearn.datasets import make_circles

# יצירת נתונים במעגלים
X_circles, y_circles = make_circles(n_samples=200, noise=0.1, factor=0.3, random_state=42)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# גרף 1 - הבעיה
axes[0].scatter(X_circles[y_circles==0, 0], X_circles[y_circles==0, 1], c='blue', s=100, label='Class 0')
axes[0].scatter(X_circles[y_circles==1, 0], X_circles[y_circles==1, 1], c='red', s=100, label='Class 1')
axes[0].set_title('❌ אי אפשר להפריד עם קו ישר!', fontsize=14)
axes[0].legend()

# ניסיון עם SVM לינארי
svm_linear = SVC(kernel='linear')
svm_linear.fit(X_circles, y_circles)
acc_linear = accuracy_score(y_circles, svm_linear.predict(X_circles))

# גרף 2 - ניסיון עם קו (נכשל)
axes[1].scatter(X_circles[y_circles==0, 0], X_circles[y_circles==0, 1], c='blue', s=100, label='Class 0')
axes[1].scatter(X_circles[y_circles==1, 0], X_circles[y_circles==1, 1], c='red', s=100, label='Class 1')

# משרטט את הקו הלינארי
w = svm_linear.coef_[0]
b = svm_linear.intercept_[0]
x_line = np.linspace(-1.5, 1.5, 100)
y_line = -(w[0] * x_line + b) / w[1]
axes[1].plot(x_line, y_line, 'g-', linewidth=2, label='Linear SVM')
axes[1].set_title(f'SVM לינארי: Accuracy = {acc_linear:.1%}', fontsize=14)
axes[1].legend()

plt.suptitle('הבעיה: נתונים לא לינאריים', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
````

<div dir="rtl" style="text-align: right; background-color: #e8f5e9; padding: 20px; border-radius: 10px; border-right: 4px solid #4caf50;">

## ✅ הפתרון: Kernel Trick!

**הרעיון:** להעביר את הנתונים למימד גבוה יותר, שם **כן** אפשר להפריד אותם עם קו!

### דוגמה פשוטה:
- נקודות על מעגל (2D) - לא ניתן להפריד בקו
- מוסיפים מימד שלישי: $z = x^2 + y^2$
- עכשיו אפשר להפריד עם **מישור** (3D)

### Kernels נפוצים:

| Kernel | שימוש | דוגמה |
|--------|-------|-------|
| **Linear** | נתונים לינאריים | הפרדה בקו ישר |
| **RBF** (ברירת מחדל) | נתונים לא לינאריים | מעגלים, צורות מורכבות |
| **Polynomial** | קשרים פולינומיאליים | עקומות |

</div>

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# השוואה: Linear vs RBF Kernel

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# הכנת meshgrid לתצוגה
x_min, x_max = X_circles[:, 0].min() - 0.5, X_circles[:, 0].max() + 0.5
y_min, y_max = X_circles[:, 1].min() - 0.5, X_circles[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                     np.linspace(y_min, y_max, 200))

# SVM לינארי
svm_linear = SVC(kernel='linear')
svm_linear.fit(X_circles, y_circles)
Z_linear = svm_linear.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

axes[0].contourf(xx, yy, Z_linear, alpha=0.3, cmap='coolwarm')
axes[0].scatter(X_circles[y_circles==0, 0], X_circles[y_circles==0, 1], c='blue', s=50, edgecolors='black')
axes[0].scatter(X_circles[y_circles==1, 0], X_circles[y_circles==1, 1], c='red', s=50, edgecolors='black')
acc_linear = accuracy_score(y_circles, svm_linear.predict(X_circles))
axes[0].set_title(f'Linear Kernel\nAccuracy: {acc_linear:.1%}', fontsize=14)

# SVM עם RBF Kernel
svm_rbf = SVC(kernel='rbf')
svm_rbf.fit(X_circles, y_circles)
Z_rbf = svm_rbf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

axes[1].contourf(xx, yy, Z_rbf, alpha=0.3, cmap='coolwarm')
axes[1].scatter(X_circles[y_circles==0, 0], X_circles[y_circles==0, 1], c='blue', s=50, edgecolors='black')
axes[1].scatter(X_circles[y_circles==1, 0], X_circles[y_circles==1, 1], c='red', s=50, edgecolors='black')
acc_rbf = accuracy_score(y_circles, svm_rbf.predict(X_circles))
axes[1].set_title(f'RBF Kernel\nAccuracy: {acc_rbf:.1%}', fontsize=14)

plt.suptitle('Kernel Trick: מלינארי ל-RBF', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

print(f"\n💡 RBF Kernel הצליח להפריד את המעגלים!")
````

## ⚙️ פרמטרים חשובים ב-SVM

---

</div>

<div dir="rtl" style="text-align: right; background-color: #fff3e0; padding: 20px; border-radius: 10px;">

### פרמטר C (Regularization)

**C** קובע את האיזון בין:
- **C גדול:** Margin קטן, פחות שגיאות על ה-Train
- **C קטן:** Margin גדול, יותר "סלחני" לשגיאות

| C | התנהגות | סיכון |
|---|---------|-------|
| **גבוה** (100) | מתאמץ לסווג נכון כל נקודה | Overfitting |
| **נמוך** (0.01) | מעדיף Margin גדול, מוותר על נקודות | Underfitting |
| **ברירת מחדל** (1) | איזון | מומלץ להתחיל |

### פרמטר gamma (ל-RBF)

**gamma** קובע כמה רחוק ההשפעה של כל נקודה:
- **gamma גבוה:** כל נקודה משפיעה רק על סביבתה הקרובה
- **gamma נמוך:** כל נקודה משפיעה רחוק

</div>

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# השפעת פרמטר C

from sklearn.datasets import make_classification

# נתונים עם קצת רעש
X_noisy, y_noisy = make_classification(n_samples=100, n_features=2, n_redundant=0, 
                                        n_informative=2, n_clusters_per_class=1,
                                        flip_y=0.1, random_state=42)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
C_values = [0.01, 1, 100]

for ax, C in zip(axes, C_values):
    svm = SVC(kernel='linear', C=C)
    svm.fit(X_noisy, y_noisy)
    
    # meshgrid
    x_min, x_max = X_noisy[:, 0].min() - 1, X_noisy[:, 0].max() + 1
    y_min, y_max = X_noisy[:, 1].min() - 1, X_noisy[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                         np.linspace(y_min, y_max, 200))
    Z = svm.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    
    ax.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
    ax.scatter(X_noisy[y_noisy==0, 0], X_noisy[y_noisy==0, 1], c='blue', s=50, edgecolors='black')
    ax.scatter(X_noisy[y_noisy==1, 0], X_noisy[y_noisy==1, 1], c='red', s=50, edgecolors='black')
    
    # Support vectors
    ax.scatter(svm.support_vectors_[:, 0], svm.support_vectors_[:, 1], 
               s=200, facecolors='none', edgecolors='black', linewidths=2)
    
    acc = accuracy_score(y_noisy, svm.predict(X_noisy))
    ax.set_title(f'C = {C}\nSupport Vectors: {len(svm.support_vectors_)}\nAccuracy: {acc:.1%}')

plt.suptitle('השפעת פרמטר C על SVM', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
````

## 📐 המתמטיקה של Perceptron

### שלב 1: סכום משוקלל
$$z = w_1 \cdot x_1 + w_2 \cdot x_2 + ... + w_n \cdot x_n + b$$

או בקיצור:
$$z = \sum_{i=1}^{n} w_i \cdot x_i + b$$

### שלב 2: פונקציית הפעלה (Activation)

**Step Function (המקורית):**
$$y = egin{cases} 1 & 	ext{if } z \geq 0 \\ 0 & 	ext{if } z < 0 \end{cases}$$

### דוגמה מספרית:
- קלטים: $x_1 = 0.5$, $x_2 = 0.3$
- משקלים: $w_1 = 0.4$, $w_2 = 0.6$
- bias: $b = -0.2$

**חישוב:**
$$z = (0.4 	imes 0.5) + (0.6 	imes 0.3) + (-0.2) = 0.2 + 0.18 - 0.2 = 0.18$$

**פלט:** $z = 0.18 \geq 0$ → $y = 1$

</div>

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# חישוב ידני של Perceptron

def perceptron(x, w, b):
    """
    x: רשימת קלטים
    w: רשימת משקלים
    b: bias
    """
    # סכום משוקלל
    z = sum(xi * wi for xi, wi in zip(x, w)) + b
    
    # פונקציית הפעלה (step)
    y = 1 if z >= 0 else 0
    
    return z, y

# דוגמה
x = [0.5, 0.3]
w = [0.4, 0.6]
b = -0.2

z, y = perceptron(x, w, b)

print("📊 חישוב Perceptron")
print("=" * 40)
print(f"קלטים (x): {x}")
print(f"משקלים (w): {w}")
print(f"bias (b): {b}")
print()
print(f"סכום משוקלל (z): {z:.2f}")
print(f"פלט (y): {y}")
print()
print("📝 פירוט החישוב:")
print(f"   ({w[0]} × {x[0]}) + ({w[1]} × {x[1]}) + ({b})")
print(f" = {w[0]*x[0]} + {w[1]*x[1]} + {b}")
print(f" = {z:.2f}")
print(f"   {z:.2f} >= 0 ? → y = {y}")
````

## 🔄 אלגוריתם הלמידה

### הרעיון:
1. התחל עם משקלים אקראיים
2. עבור על כל הדוגמאות
3. אם טעית - **עדכן את המשקלים**
4. חזור עד שאין שגיאות (או מספיק iterations)

### כלל העדכון:
$$w_i^{new} = w_i^{old} + \eta \cdot (y_{true} - y_{pred}) \cdot x_i$$

כאשר:
- $\eta$ = **Learning Rate** (קצב למידה) - כמה חזק לעדכן
- $y_{true}$ = התשובה הנכונה
- $y_{pred}$ = מה שחזינו
- $x_i$ = הקלט

### אינטואיציה:
- אם טעינו ($y_{true} 
eq y_{pred}$) → מעדכנים
- אם צדקנו ($y_{true} = y_{pred}$) → לא מעדכנים (ההפרש = 0)

</div>

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# אנימציה של Perceptron Learning

np.random.seed(42)

# נתונים פשוטים (linearly separable)
X_simple = np.array([[2, 3], [1, 1], [2, 1], [3, 2],   # Class 0
                      [5, 5], [6, 4], [5, 6], [7, 5]])  # Class 1
y_simple = np.array([0, 0, 0, 0, 1, 1, 1, 1])

# Perceptron מאפס
class SimplePerceptron:
    def __init__(self, learning_rate=0.1):
        self.lr = learning_rate
        self.weights = None
        self.bias = None
        self.history = []
    
    def fit(self, X, y, epochs=10):
        n_features = X.shape[1]
        self.weights = np.random.randn(n_features) * 0.1
        self.bias = 0
        
        for epoch in range(epochs):
            errors = 0
            for xi, yi in zip(X, y):
                # חיזוי
                z = np.dot(xi, self.weights) + self.bias
                y_pred = 1 if z >= 0 else 0
                
                # עדכון אם טעינו
                if y_pred != yi:
                    errors += 1
                    self.weights += self.lr * (yi - y_pred) * xi
                    self.bias += self.lr * (yi - y_pred)
            
            self.history.append({
                'epoch': epoch,
                'weights': self.weights.copy(),
                'bias': self.bias,
                'errors': errors
            })
            
            if errors == 0:
                break
    
    def predict(self, X):
        z = np.dot(X, self.weights) + self.bias
        return (z >= 0).astype(int)

# אימון
perceptron_model = SimplePerceptron(learning_rate=0.1)
perceptron_model.fit(X_simple, y_simple, epochs=20)

# ויזואליזציה של תהליך הלמידה
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

epochs_to_show = [0, 1, 2, 3, 4, len(perceptron_model.history)-1]
epochs_to_show = [min(e, len(perceptron_model.history)-1) for e in epochs_to_show]

for ax, epoch_idx in zip(axes, epochs_to_show):
    h = perceptron_model.history[epoch_idx]
    w = h['weights']
    b = h['bias']
    
    # נקודות
    ax.scatter(X_simple[y_simple==0, 0], X_simple[y_simple==0, 1], 
               c='blue', s=100, label='Class 0', edgecolors='black')
    ax.scatter(X_simple[y_simple==1, 0], X_simple[y_simple==1, 1], 
               c='red', s=100, label='Class 1', edgecolors='black')
    
    # קו הפרדה
    if w[1] != 0:
        x_line = np.linspace(0, 8, 100)
        y_line = -(w[0] * x_line + b) / w[1]
        ax.plot(x_line, y_line, 'g-', linewidth=2, label='Decision Boundary')
    
    ax.set_xlim(0, 8)
    ax.set_ylim(0, 8)
    ax.set_title(f'Epoch {h["epoch"]}: {h["errors"]} errors', fontsize=12)
    ax.legend(loc='upper left', fontsize=8)
    ax.set_xlabel('x₁')
    ax.set_ylabel('x₂')

plt.suptitle('Perceptron Learning: הקו משתפר עם כל epoch!', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print(f"\n✅ הלמידה הסתיימה אחרי {len(perceptron_model.history)} epochs")
print(f"   שגיאות סופיות: {perceptron_model.history[-1]['errors']}")
````

## 💡 הפתרון: שכבות נוספות!

**MLP = רשת נוירונים עם שכבה/שכבות נסתרות (Hidden Layers)**

### מבנה:
1. **Input Layer** - שכבת קלט (ה-Features)
2. **Hidden Layer(s)** - שכבות נסתרות (הקסם קורה כאן!)
3. **Output Layer** - שכבת פלט (התחזית)

### למה זה עובד?
- כל שכבה לומדת **ייצוג** אחר של הנתונים
- שכבות עמוקות יותר = דפוסים מורכבים יותר
- זה הבסיס ל-**Deep Learning**!

### פונקציות הפעלה נפוצות:

| פונקציה | תחום | שימוש |
|---------|------|-------|
| **Sigmoid** | (0, 1) | הסתברויות |
| **ReLU** | [0, ∞) | שכבות נסתרות |
| **Softmax** | (0, 1), סכום=1 | סיווג רב-קטגוריאלי |

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# ויזואליזציה של MLP

fig, ax = plt.subplots(figsize=(14, 8))
ax.set_xlim(0, 14)
ax.set_ylim(0, 10)
ax.axis('off')

# הגדרת שכבות
layers = {
    'input': {'x': 2, 'neurons': 3, 'color': 'lightblue', 'label': 'Input Layer\n(Features)'},
    'hidden1': {'x': 5, 'neurons': 4, 'color': 'lightyellow', 'label': 'Hidden Layer 1'},
    'hidden2': {'x': 8, 'neurons': 4, 'color': 'lightyellow', 'label': 'Hidden Layer 2'},
    'output': {'x': 11, 'neurons': 2, 'color': 'lightcoral', 'label': 'Output Layer\n(Classes)'}
}

def get_neuron_positions(layer_info):
    n = layer_info['neurons']
    x = layer_info['x']
    y_positions = np.linspace(2, 8, n)
    return [(x, y) for y in y_positions]

# ציור קשרים
prev_positions = None
for layer_name, layer_info in layers.items():
    positions = get_neuron_positions(layer_info)
    
    if prev_positions is not None:
        for px, py in prev_positions:
            for cx, cy in positions:
                ax.plot([px, cx], [py, cy], 'gray', alpha=0.3, linewidth=0.5)
    
    prev_positions = positions

# ציור נוירונים
for layer_name, layer_info in layers.items():
    positions = get_neuron_positions(layer_info)
    for x, y in positions:
        circle = Circle((x, y), 0.35, color=layer_info['color'], ec='black', linewidth=2)
        ax.add_patch(circle)
    
    # תווית
    ax.text(layer_info['x'], 0.8, layer_info['label'], ha='center', fontsize=10)

# כותרת
ax.text(7, 9.5, 'Multi-Layer Perceptron (MLP)', fontsize=18, fontweight='bold', ha='center')

plt.tight_layout()
plt.show()

print("📊 מבנה הרשת:")
print("   Input: 3 neurons (features)")
print("   Hidden 1: 4 neurons")
print("   Hidden 2: 4 neurons")
print("   Output: 2 neurons (classes)")
````

# 📝 סיכום המחברת

## SVM
- מחפש קו/מישור עם **Margin מקסימלי**

- **מגבלה:** יכול לפתור רק בעיות לינאריות

## MLP
- **רשת נוירונים** עם שכבות נסתרות
- פותר בעיות **לא-לינאריות** (כמו XOR)
- הבסיס ל-**Deep Learning**
- צריך **הרבה נתונים** ו-**נרמול**
