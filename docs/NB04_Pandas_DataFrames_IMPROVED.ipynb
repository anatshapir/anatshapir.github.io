````python
# Import pandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print('Pandas version:', pd.__version__)
print('\n✅ Pandas imported successfully!')
print('\n💡 Convention: We always import pandas as "pd"')
````

<div dir="rtl" style="text-align: right;">

## 📊 מבני הנתונים של Pandas

ל-Pandas יש שני מבני נתונים עיקריים:

| מבנה | תיאור | דומה ל... |
|------|-------|----------|
| **Series** | עמודה אחת (חד-ממדי) | רשימה עם אינדקס |
| **DataFrame** | טבלה (דו-ממדי) | גיליון Excel |

</div>

````python
# Series - one column
grades = pd.Series([85, 90, 78, 92, 88], name='Grades')
print('Series (single column):')
print(grades)
print(f'\nType: {type(grades)}')
````

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# DataFrame - table
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [22, 25, 23, 24, 22],
    'Grade': [85, 90, 78, 92, 88],
    'City': ['Tel Aviv', 'Haifa', 'Jerusalem', 'Tel Aviv', 'Haifa']
}
df = pd.DataFrame(data)

print('DataFrame (table):')
display(df)
print(f'\nType: {type(df)}')
````

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# Visual structure of DataFrame
import matplotlib.patches as mpatches

fig, ax = plt.subplots(figsize=(12, 6))
ax.set_xlim(0, 12)
ax.set_ylim(0, 8)
ax.axis('off')

# Title
ax.text(6, 7.5, 'DataFrame Structure', ha='center', fontsize=16, fontweight='bold')

# Column headers
headers = ['Index', 'Name', 'Age', 'Grade', 'City']
colors = ['#95a5a6', '#3498db', '#3498db', '#3498db', '#3498db']
for i, (header, color) in enumerate(zip(headers, colors)):
    rect = mpatches.Rectangle((1 + i*2, 6), 1.8, 0.7, facecolor=color, edgecolor='white', linewidth=2)
    ax.add_patch(rect)
    ax.text(1.9 + i*2, 6.35, header, ha='center', va='center', fontsize=10, color='white', fontweight='bold')

# Data rows
data_rows = [
    [0, 'Alice', 22, 85, 'Tel Aviv'],
    [1, 'Bob', 25, 90, 'Haifa'],
    [2, 'Charlie', 23, 78, 'Jerusalem'],
]
for row_idx, row in enumerate(data_rows):
    for col_idx, val in enumerate(row):
        color = '#ecf0f1' if col_idx > 0 else '#bdc3c7'
        rect = mpatches.Rectangle((1 + col_idx*2, 5.2 - row_idx*0.7), 1.8, 0.6, 
                                   facecolor=color, edgecolor='#bdc3c7', linewidth=1)
        ax.add_patch(rect)
        ax.text(1.9 + col_idx*2, 5.5 - row_idx*0.7, str(val), ha='center', va='center', fontsize=9)

# Labels
ax.annotate('Columns\n(Features)', xy=(6, 6.7), xytext=(6, 7.3),
            arrowprops=dict(arrowstyle='->', color='#3498db'), fontsize=10, ha='center', color='#3498db')

ax.annotate('Index\n(Row labels)', xy=(1.9, 4.5), xytext=(0.3, 4.5),
            arrowprops=dict(arrowstyle='->', color='#95a5a6'), fontsize=10, ha='center', color='#95a5a6')

ax.annotate('Rows\n(Samples)', xy=(10.5, 4.5), xytext=(11.5, 4.5),
            arrowprops=dict(arrowstyle='->', color='#2ecc71'), fontsize=10, ha='left', color='#2ecc71')

# Highlight one row
rect = mpatches.Rectangle((1, 4.5), 9.8, 0.6, facecolor='none', edgecolor='#2ecc71', linewidth=2, linestyle='--')
ax.add_patch(rect)

# Highlight one column
rect = mpatches.Rectangle((5, 3.2), 1.8, 3.5, facecolor='none', edgecolor='#e74c3c', linewidth=2, linestyle='--')
ax.add_patch(rect)
ax.text(5.9, 2.8, 'One Column\n(Series)', ha='center', fontsize=9, color='#e74c3c')

plt.tight_layout()
plt.show()
````

## 4.2 בחירת שורות לפי אינדקס - `iloc` ו-`loc`

- **`iloc`** - בחירה לפי מיקום (index position) - מספרים
- **`loc`** - בחירה לפי תווית (label) - שם האינדקס

</div>

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# iloc - by position (0, 1, 2, ...)
print('iloc - by position:')
print('\nFirst row (iloc[0]):')
print(df.iloc[0])
print('\nRows 0, 1, 2 (iloc[0:3]):')
display(df.iloc[0:3])
````

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# iloc with rows AND columns
print('iloc[0:3, 0:2] - first 3 rows, first 2 columns:')
display(df.iloc[0:3, 0:2])

print('\niloc[0:3, [0, 2, 4]] - first 3 rows, columns 0, 2, 4:')
display(df.iloc[0:3, [0, 2, 4]])
````

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# loc - by label
print('loc - by label:')
print('\nRow with index 0 (loc[0]):')
print(df.loc[0])
print('\nRows 0-2, columns Name and Salary:')
display(df.loc[0:2, ['Name', 'Salary']])
````

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# Visual iloc vs loc
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# iloc
ax = axes[0]
ax.set_xlim(0, 10)
ax.set_ylim(0, 8)
ax.axis('off')
ax.set_title('iloc - by Position (numbers)', fontsize=13, fontweight='bold', color='#3498db')

# Headers
for i, col in enumerate(['', 'Name', 'Age', 'Salary']):
    color = '#3498db' if i > 0 else '#95a5a6'
    rect = mpatches.Rectangle((1 + i*2, 6.5), 1.8, 0.7, facecolor=color, edgecolor='white', linewidth=2)
    ax.add_patch(rect)
    text = col if i > 0 else 'idx'
    ax.text(1.9 + i*2, 6.85, text, ha='center', va='center', fontsize=9, color='white', fontweight='bold')
    if i > 0:
        ax.text(1.9 + i*2, 7.4, f'col {i-1}', ha='center', fontsize=8, color='#3498db')

# Data
data = [['0', 'Alice', '28', '15000'], ['1', 'Bob', '35', '18000'], ['2', 'Charlie', '42', '22000']]
for row_idx, row in enumerate(data):
    ax.text(-0.2, 5.95 - row_idx*0.8, f'row {row_idx}', ha='center', fontsize=8, color='#95a5a6')
    for col_idx, val in enumerate(row):
        color = '#ecf0f1' if col_idx > 0 else '#bdc3c7'
        rect = mpatches.Rectangle((1 + col_idx*2, 5.6 - row_idx*0.8), 1.8, 0.6, 
                                   facecolor=color, edgecolor='#bdc3c7', linewidth=1)
        ax.add_patch(rect)
        ax.text(1.9 + col_idx*2, 5.9 - row_idx*0.8, val, ha='center', va='center', fontsize=9)

ax.text(5, 3, 'df.iloc[0, 1] = "Alice"', fontsize=11, ha='center', 
        bbox=dict(boxstyle='round', facecolor='#e8f4f8'))
ax.text(5, 2.3, 'df.iloc[0:2, 1:3]', fontsize=11, ha='center', 
        bbox=dict(boxstyle='round', facecolor='#e8f4f8'))

# loc
ax = axes[1]
ax.set_xlim(0, 10)
ax.set_ylim(0, 8)
ax.axis('off')
ax.set_title('loc - by Label (names)', fontsize=13, fontweight='bold', color='#9b59b6')

# Headers
for i, col in enumerate(['', 'Name', 'Age', 'Salary']):
    color = '#9b59b6' if i > 0 else '#95a5a6'
    rect = mpatches.Rectangle((1 + i*2, 6.5), 1.8, 0.7, facecolor=color, edgecolor='white', linewidth=2)
    ax.add_patch(rect)
    text = col if i > 0 else 'idx'
    ax.text(1.9 + i*2, 6.85, text, ha='center', va='center', fontsize=9, color='white', fontweight='bold')

# Data
for row_idx, row in enumerate(data):
    for col_idx, val in enumerate(row):
        color = '#ecf0f1' if col_idx > 0 else '#bdc3c7'
        rect = mpatches.Rectangle((1 + col_idx*2, 5.6 - row_idx*0.8), 1.8, 0.6, 
                                   facecolor=color, edgecolor='#bdc3c7', linewidth=1)
        ax.add_patch(rect)
        ax.text(1.9 + col_idx*2, 5.9 - row_idx*0.8, val, ha='center', va='center', fontsize=9)

ax.text(5, 3, 'df.loc[0, "Name"] = "Alice"', fontsize=11, ha='center', 
        bbox=dict(boxstyle='round', facecolor='#f3e5f5'))
ax.text(5, 2.3, 'df.loc[0:1, ["Name", "Age"]]', fontsize=11, ha='center', 
        bbox=dict(boxstyle='round', facecolor='#f3e5f5'))

plt.tight_layout()
plt.show()

print('💡 iloc uses numbers (positions)')
print('💡 loc uses labels (column names, index values)')
````

## 5.2 תנאים מרובים

אפשר לשלב כמה תנאים:
- **`&`** - וגם (AND)
- **`|`** - או (OR)
- **`~`** - לא (NOT)

⚠️ **חשוב:** כל תנאי חייב להיות בסוגריים!

</div>

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# Multiple conditions with AND
print('Age > 30 AND Salary > 18000:')
display(df[(df['Age'] > 30) & (df['Salary'] > 18000)])
````

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# Multiple conditions with OR
print('City is Tel Aviv OR City is Haifa:')
display(df[(df['City'] == 'Tel Aviv') | (df['City'] == 'Haifa')])
````

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# Using isin() for multiple values
print('City is in [Tel Aviv, Haifa] - using isin():')
display(df[df['City'].isin(['Tel Aviv', 'Haifa'])])
````

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# NOT condition
print('NOT in Engineering department:')
display(df[~(df['Department'] == 'Engineering')])
````

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# Create data with missing values
data_missing = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [28, np.nan, 42, 31, np.nan],
    'Salary': [15000, 18000, np.nan, 16000, 14000],
    'Department': ['Engineering', 'Marketing', 'Engineering', None, 'Marketing']
}
df_missing = pd.DataFrame(data_missing)

print('Data with missing values (NaN):')
display(df_missing)
````

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# Check for missing values
print('Is each value missing? (True = missing):')
display(df_missing.isna())
````

````python
# Count missing values per column
print('Missing values per column:')
print(df_missing.isna().sum())
print(f'\nTotal missing values: {df_missing.isna().sum().sum()}')
````

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# Visualize missing values
fig, ax = plt.subplots(figsize=(8, 4))
missing_counts = df_missing.isna().sum()
colors = ['#e74c3c' if x > 0 else '#2ecc71' for x in missing_counts]
bars = ax.bar(missing_counts.index, missing_counts.values, color=colors)
ax.set_ylabel('Missing Count')
ax.set_title('Missing Values by Column')
for bar, count in zip(bars, missing_counts.values):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1, 
            str(count), ha='center', fontsize=11)
plt.tight_layout()
plt.show()
````

## 7.2 אסטרטגיות לטיפול בערכים חסרים

</div>

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# Option 1: Drop rows with missing values
print('Option 1: Drop rows with ANY missing value')
df_dropped = df_missing.dropna()
display(df_dropped)
print(f'Remaining rows: {len(df_dropped)} (from {len(df_missing)})')
````

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# Option 2: Fill with a specific value
print('Option 2: Fill missing values')
df_filled = df_missing.copy()

# Fill Age with mean
mean_age = df_filled['Age'].mean()
df_filled['Age'] = df_filled['Age'].fillna(mean_age)
print(f'Filled Age with mean: {mean_age:.1f}')

# Fill Salary with median
median_salary = df_filled['Salary'].median()
df_filled['Salary'] = df_filled['Salary'].fillna(median_salary)
print(f'Filled Salary with median: {median_salary}')

# Fill Department with mode (most common)
mode_dept = df_filled['Department'].mode()[0]
df_filled['Department'] = df_filled['Department'].fillna(mode_dept)
print(f'Filled Department with mode: {mode_dept}')

print('\nResult:')
display(df_filled)
````

# חלק 8: קיבוץ וסיכום (GroupBy) 📊

---

אחת הפעולות החזקות ביותר - לסכם נתונים לפי קבוצות!

</div>

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# Back to our original data
df = pd.read_csv('employees.csv')
print('Our data:')
display(df)
````

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# GroupBy - basic
print('Average salary by Department:')
display(df.groupby('Department')['Salary'].mean())
````

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# GroupBy - multiple aggregations
print('Statistics by Department:')
dept_stats = df.groupby('Department')['Salary'].agg(['mean', 'min', 'max', 'count'])
display(dept_stats)
````

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# GroupBy - multiple columns
print('Average salary by Department AND City:')
display(df.groupby(['Department', 'City'])['Salary'].mean())
````

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# GroupBy - as DataFrame with reset_index
print('As a clean table:')
result = df.groupby('Department')['Salary'].mean().reset_index()
result.columns = ['Department', 'Avg_Salary']
display(result)
````

````python
#@title 🎨 **הדגמה** - לחצו ▶ להרצה (הקוד מוסתר)
# Visualize GroupBy result
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Average salary by department
dept_avg = df.groupby('Department')['Salary'].mean().sort_values(ascending=True)
colors = ['#3498db', '#2ecc71', '#e74c3c']
axes[0].barh(dept_avg.index, dept_avg.values, color=colors)
axes[0].set_xlabel('Average Salary')
axes[0].set_title('Average Salary by Department')
for i, v in enumerate(dept_avg.values):
    axes[0].text(v + 200, i, f'{v:,.0f}', va='center')

# Count by city
city_count = df.groupby('City').size().sort_values(ascending=True)
axes[1].barh(city_count.index, city_count.values, color='#9b59b6')
axes[1].set_xlabel('Number of Employees')
axes[1].set_title('Employee Count by City')
for i, v in enumerate(city_count.values):
    axes[1].text(v + 0.1, i, str(v), va='center')

plt.tight_layout()
plt.show()
````
